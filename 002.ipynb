### 16. ファイルをN分割する
自然数Nをコマンドライン引数などの手段で受け取り，入力のファイルを行単位でN分割せよ．同様の処理をsplitコマンドで実現せよ

split -l 1 samp.txt

```python
n = input()
if n == 0:
    print("0以外の文字を入力してください")
else:
    file_name = "data/popular-names.txt"
    with open(file_name, "r") as file:
        lines = file.readlines()
        for count, line in enumerate(lines):
            print(line, count, int(n))
            
            # N行コピーしたら次のファイルを開ける
            # countがNの倍数の時
            if count%int(n) == 0 or count == 0 : 
                print("open the file", count)
                output_file = ("data/samp/out" + str(count))
                file_out = open(output_file, "w") 

            file_out.write(line)

            # N行コピーしたら書き込んだファイルを閉じる
            # count+1がNの倍数の時
            if (count+1)%int(n) == 0 and count != 0 :
                print("close the file", count)
                file_out.close()
```

sort -k 3n popular-names.txt

```python
file_name = "data/popular-names.txt"
name_list = []
line_list = []
name_set = {}

with open(file_name, "r") as file:
    lines = file.readlines()
    for line in lines:
        l = line.split()
        name_list.append(l[0])
        line_list.append(l)
    name_set = set(name_list)
    
    # 被りなしの名前セットがどの程度出現するのか計算
    for single_name in name_set:
        count = name_list.count(single_name)
        appeared = ( count / len(name_list) ) # 出現率を計算
        
        # single_nameの出現率をlineリストに与える
        for line in line_list:
            if single_name == line[0]:
                line.insert(0, appeared)

    print(sorted(line_list))
```
